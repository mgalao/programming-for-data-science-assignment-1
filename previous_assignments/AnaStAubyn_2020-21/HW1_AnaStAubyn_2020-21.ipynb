1. Download the data.txt file

2. Open the file “data.txt” in read mode and load the data into a List

```python
data=open('C:/Users/anacs/Documents/NOVA IMS/Mestrado/Programming For Data Science/Homework_1/data.txt','r')
content=data.readlines()
L=[]
for i in content:
    L.append(float(i.replace('\n','')))
print(L)
```


3. Verify that each value in the List is a float ranging between -100 and 100;

```python
L.sort()
print(L[0],L[len(L)-1])
```


4. Round the number to the integer digit and convert each value of the List into an Integer;

```python
FINAL=[]
for i in L:
    FINAL.append(int(i))
print(FINAL)
```


5. Sort the list in descending order;

```python
FINAL.sort(reverse=1)
print(FINAL)
```


6. Save the newly sorted list of integers into a “.csv” file with the name “H1_< id >.csv”, replacing < id > by your student id.
    For example H1_20209999.csv

```python
import csv
with open('H1_2016713.csv', 'w') as hw:
    for i in FINAL:
        hw.write(str(i))
        hw.write("\n")
```

7. Print the values below -32

```python
a=-32
for i in FINAL:
    if i<a:
        print(i)
```

8. Create a dictionary with three keys. The first holds a list of the positive values from the list obtained in step 5, the second key the zeros and the last one, all the negatives.

```python
positive=[]
negative=[]
zeros=[]
for i in FINAL:
    if i <0:
        negative.append(i)
    elif i==0:
        zeros.append(i)
    else:
        positive.append(i)
d={'Positive':positive, 'Zeros':zeros, 'Negative':negative}
print(d)
```
